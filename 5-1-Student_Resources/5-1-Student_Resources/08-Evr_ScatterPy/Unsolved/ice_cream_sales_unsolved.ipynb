```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
temp = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
sales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]
scoop_price=[random.random() for value in temp]
#scoop_price=100*scoop_price
```

```python
# Tell matplotlib to create a scatter plot based upon the above data
plt.scatter(temp,sales,marker="o",facecolors="red", edgecolors="black",s=50, alpha=0.75)
plt.scatter(temp ,scoop_price,marker="*",facecolors="blue", edgecolors="black",s=sales, alpha=0.75)
```

```python
# Set the upper and lower limits of our y axis
plt.ylim(100,600)
```

```python
# Set the upper and lower limits of our x axis
plt.xlim(11.5,26)
```

```python
# Create a title, x label, and y label for our chart
plt.xlabel("Temperature(Celsius)")
plt.ylabel("Saes(dollars)")
plt.title("Ice Cream Sales VTemperature")
```
